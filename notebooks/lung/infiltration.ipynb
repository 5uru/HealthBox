```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("jonathansuru/Infiltration")
```

```python
import numpy as np
from tinygrad import Tensor, nn
from tinygrad.nn import optim
from tinygrad.nn.state import get_parameters
import time
from utils import data_loader
from resnet import ResNet
from training import train, evaluate
```

```python
labels = {
    7: 1,
    9: 0,
}
```

```python
# Load the dataset
train_shuffled_images, train_shuffled_labels = data_loader(ds["train"], labels)
test_shuffled_images, test_shuffled_labels = data_loader(ds["test"], labels)
validation_shuffled_images, validation_shuffled_labels = data_loader(ds["validation"], labels)
```

```python
model = ResNet(18, 2)
```

```python
model.load_from_pretrained()
```

```python
X_train = Tensor(train_shuffled_images.astype(np.float32))
Y_train = Tensor(train_shuffled_labels)
X_test = Tensor(test_shuffled_images.astype(np.float32))
Y_test = Tensor(test_shuffled_labels)
```

```python
lr = 5e-3
```

```python
for _ in range(5):
    optimizer = optim.SGD(get_parameters(model), lr=lr, momentum=0.9)
    train(model, X_train, Y_train, optimizer, 100, BS=32)
    evaluate(model, X_test, Y_test, num_classes=2)
    lr /= 1.2
    print(f'reducing lr to {lr:.7f}')
```
